### Extracting 2% of total dataset

```r
diabetes_sub <- diabetes[sample(1:nrow(diabetes), 0.02*(253680), replace=FALSE),]
```

```r
sprintf("The dataset has %s records with a total of %s variables",nrow(diabetes_sub),
ncol(diabetes_sub))
```

### Prepare the data for Analysis

#### Handling Missing Values

```r
sprintf("The dataset contain %s number of missing values.", sum(is.na(diabetes_sub)))
```

#### Normalization of data

```r
diabetes_sub_norm <- as.data.frame(lapply(diabetes_sub[,2:ncol(diabetes_sub)], scale))
```

#### Checking on association

```r
# Step 1: Create a correlation matrix
corr_matrix <- cor(diabetes_sub_norm)

# Step 2: Visualize the correlation matrix using corrplot
install.packages("corrplot")
library(corrplot)  # Make sure to load the corrplot package if you haven't already
corrplot(corr_matrix, type = "lower", method = "square")
```
