## **Practice II: Divise Clustering Algorithm**

### **Installing Necessary packages**

```r
install.packages("tidyverse")
install.packages("factoextra")
```

### **Loading Necessary packages**

```r
library(tidyverse) # for data manipulation
library(cluster) # for clustering algorithms
library(factoextra) # for cluster visualization
library(dendextend) # for comparing two dendrograms
```

### **Loading the dataset**

```r
diabetes <- read.csv('Data/diabetes.csv')
```

### **Sampling Dataset**

Create a sample of the dataset i.e. 2% of the total dataset due to the computing power of the devices

```r
diabetes_sub <- diabetes[sample(1:nrow(diabetes),0.02*(253680), replace = FALSE),]
```

### **Normalizing Dataset**

Normalize the dataset to avoid challenges of large values vs small values

```r
diabetes_sub_norm <- as.data.frame(lapply(diabetes_sub[,2:ncol(diabetes_sub)], scale))
```

### **Clustering Dataset**

Cluster the normalized dataset using the diana() function (i.e. divisive analysis function)

```r
dv <- diana(diabetes_sub_norm, metric ="manhattan", stand = TRUE)
```

### **Visualisation**

Visualize the clusters formed

```r
plot(dv)
```

Cut into 2 groups

```r
dv2 <- cutree(as.hclust(dv), k=2)
table(dv2)
```

Display the sample output of the groups formed

```r
rownames(diabetes_sub_norm) [dv2 == 1]
```
